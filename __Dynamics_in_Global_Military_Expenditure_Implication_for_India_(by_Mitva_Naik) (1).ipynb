```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
##Loading SIPRI global Military expenditure dataset, excel file##
df=pd.read_excel("/content/SIPRI-Milex-data-1949-2024_2 OG.xlsx")
```

```python
df=pd.read_excel("/content/SIPRI-Milex-data-1949-2024_2 OG.xlsx",sheet_name="Constant (2023) US$", skiprows=5)
```

```python
display(df)
```

```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

```python
df = df.drop('Notes', axis=1)
df = df.iloc[1:].reset_index(drop=True)
display(df)
```

```python
df=df.drop("Unnamed: 1", axis=1)
display(df)
```

```python
year_columns = df.columns.drop(['Country'])
```

```python
df[year_columns] = df[year_columns].apply(pd.to_numeric, errors='coerce')
```

```python
display(df.info())
```

```python
df.shape
```

```python
df_long = df.melt(id_vars='Country', var_name='Year', value_name='Expenditure')     ##formating table into long format ##
display(df_long.head())
```

```python
df_long.to_csv('cleaned_military_expenditure_by country_constant US$.csv', index=False)   ##converted cleaned dataset into csv file##
```

```python
df_last10_years=df_long[df_long["Year"]>= 2014]
df_last10_years["Expenditure"]=df_last10_years["Expenditure"]*1000000
```

```python
df_last10_years.to_csv("last10_years_military expenditure_by_country.csv", index=False)
```

```python
#Global Trend ##

global_trend = df_last10_years.groupby("Year")["Expenditure"].sum().reset_index()

plt.figure(figsize=(10,6))
plt.plot(global_trend["Year"], global_trend["Expenditure"], marker="o")
plt.title("Global Trend of Military Expenditure over the last 10 years")
plt.xlabel("Year")
plt.ylabel("USD  (2023 prices)")
plt.grid(True)
plt.show()

```

```python
#Top 10 spenders by latest years##

latest_year=df_last10_years["Year"].max()
top10=(df_last10_years[df_last10_years["Year"]==latest_year].sort_values(by="Expenditure", ascending=False).head(10))

fig=px.bar(top10, x="Country", y="Expenditure", title="Top 10 Military spenders in latest year(2024)", labels= {"Expenditure": "USD"} )
fig.show()
```

```python
##Inida vs. peers##

countries=["India", "United States of America",  "China", "Russia", "Pakistan"]

df_selected=df_last10_years[df_last10_years["Country"].isin(countries)]

fig=px.line(df_selected, x="Year", y="Expenditure", color="Country", title="India vs. Peers- Military Spending")
fig.show()
```

```python
## Loading world bank database % GDP##

wb= pd.read_csv("/content/API_MS.MIL.XPND.GD.ZS_DS2_en_csv_v2_548570.csv", skiprows=4)
```

```python
wb.head(10)
```

```python
wb.info()
```

```python
wb=wb.melt(id_vars=["Country Name"], var_name= "Year", value_name="%GDP")
```

```python
wb["Year"]=pd.to_numeric(wb["Year"], errors="coerce")
wb=wb[wb["Year"] >= 2014]
```

```python
## Top 10 avg. of %GDP##

wb_avg= (wb.groupby("Country Name")["%GDP"].mean().reset_index())

print("Top 10 countries by Military spending by % GDP ")
print(wb_avg.sort_values(by="%GDP", ascending=False).head(10))
```

```python

top10_gdp = wb_avg.sort_values(by="%GDP", ascending=False).head(10)

fig = px.bar(top10_gdp,
             x="Country Name",
             y="%GDP",
             title="Top 10 Countries for Military Expenditure by avg % of GDP (2014-2023)",
             labels={"Country Name": "Country", "%GDP": "Average % of GDP"})

fig.show()
```
